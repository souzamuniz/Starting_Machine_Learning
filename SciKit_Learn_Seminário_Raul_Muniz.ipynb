# Abordando Conceitos Teóricos

1.   **O que é Machine Learning ?**

*   Machine Learning é um campo de estudo que dá ao computador a habilidade de aprender sem ser explicitamente programado (Arthur Samuel, 1959)

2.   **Formas/Tipos do Aprendizado de Máquina**


*   Aprendizado Supervisionado


>> No Aprendizado Supervisionado os dados contém as soluções desejadas. Estas são chamadas de "labels"
>> Algum exemplos de algorítmos mais famosos dessa classe são :  
*     k-Nearest Neighbors
*     Linear Regression
*     Logistic Regression
*     Support Vector Machines (SVMs)
*     Decision Trees and Random Forests
*     Neural networks


*   Aprendizado Não Supervisionado
>> Os dados neste tipo de algorítmo não possuem labels. Os algoritmos encontram as relações sem nenhuma ajuda do programador.
>> Algum exemplos de algorítmos mais famosos dessa classe são : 
    *    Clustering
    *    K-Means
    *    DBSCAN
    *    Hierarchical Cluster Analysis (HCA)



*   Aprendizado Semi Supervisionado
>> Alguns dados neste tipo de algorítmo possuem labels e outros não. Tem-se que os dados são do tipo "partially labeled training data" . Os algoritmos encontram as relações sem nenhuma ajuda do programador. Aqui temos uma combinação de supervised e unsupervised learning

>> Exemplo: Some photo-hosting services, such as Google Photos, are good examples of this. Once
you upload all your family photos to the service, it automatically recognizes that the
same person A shows up in photos 1, 5, and 11, while another person B shows up in
photos 2, 5, and 7. This is the unsupervised part of the algorithm (clustering). Now all
the system needs is for you to tell it who these people are. Just one label per person,4
and it is able to name everyone in every photo, which is useful for searching photos

*   Aprendizado por Reforço
>> O Prendizado por reforço é um pouco mais diferente dos outros métodos. Ele aprende por meio de reforço, isto é, a partir de uma tomada de decisão que o algorítmo toma, ele poderá acertar ou errar. Caso ele erre, sofrerá uma punição, já se acertar receberá uam recompensa. Dessa forma ele chegará a uma melhor "policy" que trará mais recompensas ao longo do tmepo









## Testagem e Validação
Como fazer para avaliar o nosso modelo ?

Ídeia Principal--> Dividir nossos dados em dois datasets, em um nós iremos usar para treinar e formar nosso modelo (training set) e o outro usaremos para testar o nosso modelo (test set)

# Problema de Classificação vs Problema de Regressão 
A *Classificação* estará trabalhando com variáveis discretas/qualitativas, enquanto a *Regressão* com variáveis contínuas. Logo, poderemos encarar uma situação de duas maneiras...

Supondo que nosso estudo seja estimar a temperatura de amanhã. Caso utilziemos uma abordagem de classificação iremos estimar se a temperatura futura será fria, quente ou normal. Já, se vamos utilizar a regrssão, tentaremos dar um valor a temperatura de amanhã, 25 graus, por exemplo.

```python
#Importando algumas bibliotecas para o primeiro exemplo
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import numpy as np
```

```python
# Gerando dados para Análise

Altura = [1.26, 1.65, 1.8 , 1.86, 1.05, 1.41]
Peso = [31.71, 58.89, 67.95, 90.6 , 22.65, 54.36]

df = pd.DataFrame({'Altura (m)':Altura, 'Peso (Kg)':Peso})
df
```

```python
# Visualizando os dados

plt.scatter(df['Altura (m)'], df['Peso (Kg)'])
plt.xlabel('Altura (m)')
plt.ylabel('Peso (Kg)')
plt.title('Altura vs Peso ')

plt.show()
```

```python
# A ideia, agora, é desenvolver um modelo que consiga predizer o peso por meio da análse do valor da altura

# from sklearn.linear_model import LinearRegression  #Importação da regressão linear pelo scikitlearn

X_train = df['Altura (m)'].to_numpy().reshape(-1,1)
y_train = df['Peso (Kg)'].to_numpy().reshape(-1,1)
```

```python
# Veja como o reshape funciona... Ele é capaz de transformar nossos dados em um vetor coluna com 'n' linhas. Assim formulamos um X_train

#Formato Inicial do X_train
df['Altura (m)'].to_numpy()

# #Formato Final
X_train
```

Fonte da imagem:https://stackoverflow.com/questions/51150153/valueerror-expected-2d-array-got-1d-array-instead

image.png

```python
# Construindo nosso modelo
linear_regressor = LinearRegression()
linear_regressor.fit(X_train,y_train) # Aplicação da regressão linear, treinando com os parametros X_train e y_train
```

```python
# Depois de Construir o modelo, resta ver os resultados das previsões do modelo

#1º vamos Criar dados inputs para gerarmos a previsão
valores_input_do_modelo = np.linspace(0.9,2)

#2º Vamos prever os valores de acordo com os novos inputs
Y_pred = linear_regressor.predict(valores_input_do_modelo.reshape(-1, 1))
```

```python
#Agora Vamos Visualizar os dados e os valores previstos
plt.scatter(df['Altura (m)'], df['Peso (Kg)'], label='Dados')
plt.plot(valores_input_do_modelo, Y_pred, color='k', label='Modelo Linear')
plt.xlabel('Altura (m)')
plt.ylabel('Peso (Kg)')
plt.title('Altura vs Peso ')
plt.legend()
plt.show()
```

```python
#Resumindo todos os passos a uma célula

# A ideia, agora, é desenvolver um modelo que consiga predizer o peso por meio da análse do valor da altura

# from sklearn.linear_model import LinearRegression  #Importação da regressão linear pelo scikitlearn

X_train = df['Altura (m)'].to_numpy().reshape(-1,1)
y_train = df['Peso (Kg)'].to_numpy().reshape(-1,1)


linear_regressor = LinearRegression()
linear_regressor.fit(X_train,y_train)

valores_input_do_modelo = np.linspace(0.9,2)

Y_pred = linear_regressor.predict(valores_input_do_modelo.reshape(-1, 1))

plt.scatter(df['Altura (m)'], df['Peso (Kg)'], label='Dados')
plt.plot(valores_input_do_modelo, Y_pred, color='k', label='Modelo Linear')
plt.xlabel('Altura (m)')
plt.ylabel('Peso (Kg)')
plt.title('Altura vs Peso ')
plt.legend()
plt.show()
```

# Continuação do Aprendizado Supervisionado com o SciKit-Learn

### KNN (k-Nearest Neighbors) --> O classificador KNN  utiliza medidas de distância entre os pontos já classificados e o ponto a ser classificado para teterminar a classe do mesmo, onde o parâmetro conhecido como tamanho da vizinhança estabelece o número de pontos próximos considerados.

```python
# Pegando dados da internet do github 

df = pd.read_csv('https://raw.githubusercontent.com/ronaldolagepessoa/machine-learning/master/dados/exemplo2.csv')
df.head()
```

```python
#Vamos visualizar nossos dados por meio de um gráfico scatter

plt.figure(figsize=(14, 10))
plt.scatter(df[df.risco == 'ruim']['idade'], df[df.risco == 'ruim']['conta_corrente'])
plt.scatter(df[df.risco == 'bom']['idade'], df[df.risco == 'bom']['conta_corrente'])
plt.xlabel('Idade')
plt.ylabel('Conta')
plt.legend(['Ruim', 'Bom'])
```

```python
# Vale lembrar que estamos trabalhando com Supervised Learning. Isso quer dizer que temos os valores dos labels. Contudo, é necessário separálos. Assim...

X_data_set = df.drop('risco', axis=1)
y_data_set = df.risco
```

##Relembrando...
>>>##### Testagem e Validação
>>>##### Como fazer para avaliar o nosso modelo ?
>>>##### Ídeia Principal--> Dividir nossos dados em dois datasets, em um nós iremos usar para treinar e formar nosso modelo (training set) e o outro usaremos para testar o nosso modelo (test set) 

```python
#  Agora vamos separar de todos os dados que temos  uma parcela para ser treinada e outra parcela para testarmos nosso modelo e assim avaliá-lo.
# Para isso o SciKit Learn tem uma função específica que é MUITO utilizada chamada: train_test_split

from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

X_train, X_test, y_train, y_test = train_test_split(X_data_set.values, y_data_set, test_size= 0.75, random_state=42) # Costuma-se utilizar o tamanho da amostra para treino sendo 75% e os 25% restante ficam para o teste e validação do modelo. O random_state serve para  fixar a aleatoriedade da divisão dos conjuntos e deve ser qualquer número inteiro
# Se não especificar que quer os valores do X_train ele da um warining no final da predição

```

***Curiosidade:*** Many students and practitioners use this number(42) as random state is because it is used by many instructors in online courses. They often set the random state or numpy seed to number 42 and learners follow the same practice without giving it much thought.

To be specific, 42 has nothing to do with AI or ML. It is actually a generic number, In Machine Learning, it doesn't matter what the actual random number is, as mentioned in scikit API doc, any INTEGER is sufficient enough for the task at hand.

### Importancia de dividir os dados em teste e treino...
Caso  os dados sejam todos utilziados para treino do modelo poderá ocorrer overfiting.
Os dados de teste são uma parcela dos labels guardadas para futura comparação com os labels gerados pelo modelo 

### Criando o nosso modelo com o KNN
O objetivo do classificador será a determinação do risco de uma novo cliente, baseado na idade e valor em conta corrente do mesmo.
O classificador KNN está presente no pacote sklearn no caminho sklearn.neighbors e é identificado como KNeighborsClassifier.

```python
# Importando 
from sklearn.neighbors import KNeighborsClassifier
```

```python
# Chamando a classe, inserindo os parâmetros e atribuindo-a em uma variável

knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train) 
ac_sem_norm = accuracy_score(y_test, knn.predict(X_test))
ac_sem_norm
```

```python
# Agora vamos pegar valores aleatórios para inserir como input no nosso modelo e prever o risco do novo cliente
knn.predict([[18, 500]])
```

IMPORTÂNCIA DA NORMALIZAÇÃO

*   Gera pesos mais regulares entre as variáveis estudadas

Existem vários tipos de normalizadores, Para o exemplo, aplicaremos o normalizador MinMaxScaler presente no SCIKIT-LEARN, que reformula os dados para um intervalo entre 0 e 1. Criaremos um objeto da classe MinMaxScaler chamado normalizador e utilizaremos o comando MinMaxScaler.fit_transform() para ajustar o normalizador aos dados e executar a transformação do mesmo, passando como parâmetro os inputs X.





```python
from sklearn.preprocessing import MinMaxScaler

normalizador = MinMaxScaler()
X_norm = normalizador.fit_transform(X_data_set.values)
X_norm
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_norm, y_data_set, test_size=0.75, random_state=42)

knn_norm = KNeighborsClassifier(n_neighbors=5)
knn_norm.fit(X_train, y_train)

ac_com_norm = accuracy_score(y_test, knn_norm.predict(X_test))
ac_com_norm
```

```python
# APENAS POR TRATAR NOSSO DADO PREVIAMENTE COM UMA NORMALIZAÇÃO HOUVE UMA MELHORIA SUBSTANCIAL NA RESPOSTA DO NOSSO MODELO
print('Com normalização: ',ac_com_norm)
print('Sem normalização: ',ac_sem_norm)
```

```python
# Fazendo Previsão com Dados de um suposto novo cliente
X_new = [[18, 2000]]
X_new = normalizador.transform(X_new)
clas_do_cliente = knn_norm.predict(X_new)
clas_do_cliente[0]
```
