## Criando nossa Rede Neural MLP

```python
# Vamos definir nossos dados para treino sendo as imagens (que são matrizes)
imagens = digits.images
gaba = digits['target'] #gaba = gabarito

index = 1  # index para o teste ser visto

plt.figure(figsize=(2, 2))
plt.imshow(imagens[index], cmap=plt.cm.gray_r)
plt.title('Gabarito: %i' % gaba[index])
plt.show()
```

```python
# Vamos transformar nossas matrizes em um vetor coluna, para isso utilizaremos o comando reshape:
dados = imagens.reshape((len(imagens), -1))
```

```python
# Verificando a primeira Imagem
imagens[0]
```

```python
# A primeira Imagem em forma de vetor coluna
dados[0]
```

```python
# Uma outra forma de normalizar é dividir o valor pelo maior valor  encontrado, garantindo que eles ficarão entre os valores de zero e um:
dados = dados/15. # Transforma el rango 0-15 en un rango 0-1.
```

```python
# O primeiro dado normnalizado é:
dados[0]
```

```python
# Criando a MLP (Multi-Layer Perceptron)
# A documentação sobre cada parâmetro pode ser encontrada de maneira mais completa em: https://scikit-learn.org/stable/modules/generated/sklearn.neural_network.MLPClassifier.html

from sklearn.neural_network import MLPClassifier
mlp = MLPClassifier(activation='relu', hidden_layer_sizes=(64,), 
                    max_iter=100, learning_rate_init=.1)


# Vamos dividir manualmente para entender o processo 

# Treinando a rede com uma quantidade n de dados (trainning dataset)
n = 898
mlp.fit(dados[:n], gaba[:n])

# Dados qu eutilizaremos para verificar as predições do nosso modelo (test dataset)
esperado = gaba[n:]
predicao = mlp.predict(dados[n:])

# resultados.
from sklearn import metrics
print("Reporte del clasificador: \n %s\n %s\n" 
      %(mlp, metrics.classification_report(esperado, predicao)))
print("Matriz de confusión:\n%s" % metrics.confusion_matrix(esperado, predicao))
```

La documentación sobre las métricas de sklearn se encuentra aquí:
